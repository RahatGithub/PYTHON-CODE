```python
# Function to print the 
def print_board():
    for i, row in enumerate(board):
        print(" | ".join(row))
        if not i==2: print("-" * 9) 



# Function to check if the board is full
def is_board_full():
    for row in board:
        for cell in row:
            if cell == ' ':
                return False
    return True


# Function to check if the move can lead to a winning step
def check_winning_move(row, col, player):
    # Check for horizontal match
    if board[row][(col + 1) % 3] == board[row][(col + 2) % 3] == player:
        return True
    # Check for vertical match
    if board[(row + 1) % 3][col] == board[(row + 2) % 3][col] == player:
        return True
    # Check for diagonal match 
    if row == col or row + col == 2:
        if board[(row + 1) % 3][(col + 1) % 3] == board[(row + 2) % 3][(col + 2) % 3] == player:
            return True
        if board[(row + 1) % 3][(col + 2) % 3] == board[(row + 2) % 3][(col + 1) % 3] == player:
            return True
    return False


# Get the user's move
def get_user_move():
    while True:
        row, col = map(int, input("Your turn: ").split(" "))
        if 0 <= row < 3 and 0 <= col < 3 and board[row][col] == ' ':
            return row, col
        else:
            print("INVALID MOVE!")


# Play a turn 
def play_a_turn(player): 
    # user's turn
    if player == 'O':
        row, col = get_user_move()
        return [row,col] 
    # agent's turn
    else: 
        for row in range(3):
            for col in range(3):
                if board[row][col] == ' ':
                    board[row][col] == player
                    print(f"Agent's turn: {row} {col}")
                    return [row,col] 


# main function 
def main():
    global board
    board = [[' ' for _ in range(3)] for _ in range(3)]
    players = ['X', 'O']
    count = 0
    while True:
        player = players[count%2]
        turn = play_a_turn(player)
        row, col = turn[0], turn[1]
        board[row][col] = player
        print_board()
        winner = check_winning_move(row, col, player)
        if winner:
            if player == 'X': print("The agent won")
            elif player == 'O': print("Congrats! You won!")
            break # we've got the winner
        elif is_board_full():
            print("It's a draw...")
            break
        count+=1

main()
```
